## Gradient Descent in 1D

```python
# Create two functions, for function and its derivative
def fx(x):
  return 3*x**2 - 3*x + 4

# derivative of the function
def deriv(x):
  return 6*x - 3
```

```python
# Plot both functions over a range of values
x = np.linspace(-2,2,2001)

plt.plot(x, fx(x), x, deriv(x))
plt.xlim([-2,2])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["y", "dy"])
plt.show()
```

```python
# Lets define a random starting point
localmin = np.random.choice(x,1)
print(localmin)

# define learning parameters
learning_rate = 0.01
training_epochs = 100

# Run through a training session
for i in range(training_epochs):
  # take the derivative
  grad = deriv(localmin)
  # update the value of the local minima
  localmin = localmin - learning_rate*grad

print(localmin)
```

```python
 # Plot the results
plt.plot(x, fx(x), x, deriv(x))
plt.plot(localmin, deriv(localmin),"ro")
plt.plot(localmin, fx(localmin), "ro")

plt.xlim([-2,2])
plt.grid()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend(["f(x)", "df", "f(x) min"])
plt.title("Emipirical local minimum is {:.3f}".format(localmin[0]))
plt.show()
```

```python
# Lets store the parameters
localmin = np.random.choice(x,1)
modelparams = np.zeros((training_epochs,2))

for i in range(training_epochs):
  grad = deriv(localmin)
  localmin = localmin - learning_rate*grad
  modelparams[i,:] = localmin, grad

modelparams[:5,:]
```

```python
# Plot the gradient over iterations
fig,ax = plt.subplots(1,2,figsize=(12,5))

for i in range(2):
  ax[i].plot(modelparams[:,i], "o-")
  ax[i].set_xlabel("Iteration")
  ax[i].set_title("Final Estimated Minimum: {:.3f}".format(localmin[0]))

ax[0].set_ylabel("Local Minimum")
ax[1].set_ylabel("Derivative")

plt.show()
```

## Gradient Descent in 2D

```python
# We will use the peaks function, he logo of matlab
def peaks(x,y):
  # expand to a 2D mesh
  x,y = np.meshgrid(x,y)
  z = (3*(1-x)**2*np.exp(-(x**2)-(y+1)**2) 
      - 10*(x/5-x**3-y**5)*np.exp(-x**2-y**2)
      - 1/3*np.exp(-(x+1)**2-y**2))
  return z
```

```python
# create the landscape
x = np.linspace(-3,3,201)
y = np.linspace(-3,3,201)

Z = peaks(x,y)

plt.imshow(Z, extent=[x[0],x[-1],y[0],y[-1]],vmin=-5, vmax=5, origin="lower")
plt.colorbar()
plt.show()
```

```python
# Derivtives with Sympy
sx,sy = sym.symbols("sx,sy")

sZ = (3*(1-sx)**2*sym.exp(-(sx**2)-(sy+1)**2) 
      - 10*(sx/5-sx**3-sy**5)*sym.exp(-sx**2-sy**2)
      - 1/3*sym.exp(-(sx+1)**2-sy**2))

# Create functions from sympy-computed derivative
df_x = sym.lambdify((sx,sy), sym.diff(sZ,sx), "sympy")
df_y = sym.lambdify((sx,sy), sym.diff(sZ,sy), "sympy")

df_x(1,1).evalf()
```

```python
# Pick a andom startig point between -2 to 2
localmin = np.random.rand(2)*4 - 2
startpoint = localmin[:] # make a copy without reasigning

# define the learning parameters
learning_rate = 0.01
training_epochs = 100

trajectory = np.zeros((training_epochs,2))

for i in range(training_epochs):
  grad = np.array([df_x(localmin[0], localmin[1]).evalf(),
                   df_y(localmin[0], localmin[1]).evalf()])
  localmin = localmin - learning_rate*grad
  trajectory[i,:] = localmin

print(localmin)
print(startpoint)
```

```python
plt.imshow(Z, extent=[x[0],x[-1],y[0],y[-1]],vmin=-5, vmax=5, origin="lower")
plt.plot(startpoint[0], startpoint[1], "bs")
plt.plot(localmin[0], localmin[1], "ro")
plt.plot(trajectory[:,0], trajectory[:,1], "r")
plt.legend(["rnd start", "local min"])
plt.colorbar()
plt.show()
```

### Run Parametric Experiments

#### Experiment 1: Varying Starting Points

```python
# Experiment 1: Systematically Vary the starting locaions
x = np.linspace(-2*np.pi, 2*np.pi, 401)
localmin = np.random.choice(x,1)

# learning parameters
learning_rate = 0.01
training_epochs = 1000

startlocs = np.linspace(-5,5,50)
finalres = np.zeros(len(startlocs))

# Loop over the stating points
for idx,localmin in enumerate(startlocs):

  # Training
  for i in range(training_epochs):
    grad = deriv(localmin)
    localmin = localmin - grad*learning_rate
  
  # Store the final value
  finalres[idx] = localmin

# Plot
plt.plot(startlocs, finalres, "s-")
plt.xlabel("Starting Guess")
plt.ylabel("Final Guess")
plt.plot()
```

## Code Challenge: Fixed vs Dynamic Learning Rate

Explore ways of changing the learning rate: i.e., it will change at each iteration of the GD.

```python
# We will use the simple function of 1D GD
x = np.linspace(-2,2,2001)

def fx(x):
    return 3*x**2 - 3*x + 4

def deriv(x):
    return 6*x - 3

plt.plot(x,fx(x), x, deriv(x))
plt.xlim([-2,2])
plt.legend(["f(x)", "df"])
plt.grid()
plt.show()
```

### GD with fixed learning rate

```python
localmin = np.random.choice(x,1)
print(localmin)

initval = localmin[:]

learning_rate = 0.01
training_epochs = 50

modelparamsFixed = np.zeros((training_epochs, 3))

for i in range(training_epochs):

    # gradient
    grad = deriv(localmin)

    # non-adaptive learning rate (set-up for lr modification)
    lr = learning_rate

    # update the parameter
    localmin = localmin - grad*lr

    # store
    modelparamsFixed[i,:] = localmin, grad, lr
```

### Gradient-Based Learning Rate

```python
localmin = np.random.choice(x,1)
print(localmin)

initval = localmin[:]

learning_rate = 0.01
training_epochs = 50

modelparamsGrad = np.zeros((training_epochs, 3))

for i in range(training_epochs):

    # gradient
    grad = deriv(localmin)

    # Scale the learning rate by the gradient
    lr = learning_rate*np.abs(grad)

    # update the parameter
    localmin = localmin - grad*lr

    # store
    modelparamsGrad[i,:] = localmin, grad, lr
```

### GD Dynamic: Time-based Learning Rate

```python
# redefine the parameters
localmin = np.random.choice(x,1)
print(localmin)

initval = localmin[:]

training_epochs = 50

learning_rate = 0.1
localmin = initval

# Run through training and store all the results
modelparamsTime = np.zeros((training_epochs, 3))

for i in range(training_epochs):
    
    grad = deriv(localmin)
    
    # lr will decrease as learning epochs increase
    lr = learning_rate*(1-(i+1)/training_epochs)

    localmin = localmin - lr*grad

    modelparamsTime[i,:] = localmin, grad, lr
```

```python
# Plot
fig,ax = plt.subplots(1,3, figsize=(10,3))

for i in range(3):
    ax[i].plot(modelparamsFixed[:,i], "o-", markerfacecolor="w")
    ax[i].plot(modelparamsGrad[:,i], "o-", markerfacecolor="w")
    ax[i].plot(modelparamsTime[:,i], "o-", markerfacecolor="w")
    ax[i].set_xlabel("Iteration")

ax[0].set_ylabel("Local Minimum")
ax[1].set_ylabel("Derivative")
ax[2].set_ylabel("Learning Rate")
ax[2].legend(["Fixed lr", "Grad-Based lr", "Time-Based lr"])

plt.tight_layout()
plt.show()
```
